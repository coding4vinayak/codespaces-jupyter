multiplal chain

```python
capital_templete = PromptTemplate(input_variables=["country"],
template = "Please tell me the capital of the {country}")
capital_chain=LLMChain(llm=llm_huggingface, prompt=capital_templete)

famous_templete = PromptTemplate(input_variables=['capital'],
template="suggest me some amazing places to visit in {capital}"                               )
```

```python
famous_chain = LLMChain(llm=llm_huggingface, prompt=famous_templete )
```

```python
from langchain.chains import SimpleSequentialChain
chain = SimpleSequentialChain(chains=[capital_chain, famous_chain])
chain.run("india")
```
